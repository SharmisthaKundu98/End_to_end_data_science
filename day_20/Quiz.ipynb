# Quiz 1

1. What is Flask?

   - A. A Python library for developing web applications
   - B. A JavaScript library for developing web applications
   - C. A Ruby library for developing web applications
   - D. None of the above

   **Answer: A**
   
   **Explanation:** Flask is a micro web framework written in Python that provides a simple and easy way to build web applications.

2. What is the main feature of Flask?

   - A. Lightweight and simple to use
   - B. Robust and complex to use
   - C. Heavyweight and simple to use
   - D. Heavyweight and complex to use

   **Answer: A**
   
   **Explanation:** Flask is designed to be a lightweight and simple to use framework for building web applications.

3. What is the main advantage of using Flask over a full-stack framework like Django?

   - A. Flask is more secure than Django.
   - B. Flask has a simpler learning curve compared to Django.
   - C. Flask is more scalable than Django.
   - D. None of the above

   **Answer: B**
   
   **Explanation:** Flask is a micro framework with a simpler and more straightforward design compared to full-stack frameworks like Django, making it easier for beginners.

4. What is Flask in Python used for?

   - A. Data Visualization
   - B. Gaming Development
   - C. Machine Learning
   - D. Web Development

   **Answer: D**
   
   **Explanation:** Flask is a micro web framework in Python used for web development, allowing developers to build and deploy web applications.

5. What is app routing in Flask?

   - A. It is a way to handle user authentication in Flask.
   - B. It is a method to store and retrieve data from a database in Flask.
   - C. It is a process of mapping URLs to Python functions in a Flask application.
   - D. It is a way to create and manage sessions in Flask.

   **Answer: C**
   
   **Explanation:** App routing in Flask is the process of defining the URLs that map to specific functions or methods in the application.

6. What is the correct syntax to define a URL route in Flask?

   - A. `@app.route('/')`
   - B. `app.route('/')`
   - C. `app.add_route('/')`
   - D. None of the above

   **Answer: A**
   
   **Explanation:** In Flask, the correct syntax to define a URL route is by using the `@app.route()` decorator.

7. How do you specify a URL parameter with a default value in Flask?

   - A. `@app.route('/')`
   - B. `@app.route('/=default')`
   - C. `@app.route('/', defaults={'param': 'default'})`
   - D. None of the above

   **Answer: C**
   
   **Explanation:** To specify a URL parameter with a default value in Flask, you can use the "defaults" argument of the `@app.route()` decorator.
