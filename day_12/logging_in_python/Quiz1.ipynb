# Quiz 1

### 1. To open a file `c:\scores.txt` for reading, we use:

- `infile = open(“c:\scores.txt”, “r”)`
- `infile = open(“c:\\scores.txt”, “r”)`
- `infile = open(file = “c:\scores.txt”, “r”)`
- `infile = open(file = “c:\\scores.txt”, “r”)`

**Answer:** `infile = open(“c:\\scores.txt”, “r”)`

**Explanation:** Execute `help(open)` to get more details.

### 2. Which of the following statements are true?

- When you open a file for reading, if the file does not exist, an error occurs
- When you open a file for writing, if the file does not exist, a new file is created
- When you open a file for writing, if the file exists, the existing file is overwritten with the new file
- All of the mentioned

**Answer:** All of the mentioned

**Explanation:** The program will throw an error.

### 3. Which one of the following is not attributes of file?

- closed
- softspace

### 5. Which of these methods are used to read from a file?

- get()
- read()
- scan()
- readFileInput()

**Answer:** read()

**Explanation:** Each time `read()` is called, it reads a single byte from the file and returns the byte as an integer value. `read()` returns -1 when the end of the file is encountered.

### 6. The logging …….. in Python.

- function
- module
- variable
- datatype

**Answer:** module

**Explanation:** The logging module in Python is a ready-to-use and powerful module that is designed to meet the needs of beginners as well as enterprise teams.

### 7. Which one of the following is the lowest level of logging?

- debug
- warning
- error
- none of the above

**Answer:** debug

**Explanation:** Debug is the lowest logging level, used for logging diagnostic information about the application.

### 8. Which of the following is not a logging function in Python?

- logger
- filter
- critical
- All of the above

**Answer:** filter

**Explanation:** "filter" is not a logging function; it is a logging concept used to control the flow of log messages.

### 9. Which of these is not involved in the process of debugging?

- Fixing
- Isolating
- Identifying
- Testing

**Answer:** Testing

**Explanation:** Testing is a separate process from debugging.

### 10. What is the purpose of the import statement in Python?

- import the statement into a Python program
- import the function into a Python program
- import the module into a Python program
- none of the above

**Answer:** import the module into a Python program

**Explanation:** The import statement is used to import a module into a Python program, allowing access to its definitions and statements.

### 11. How can you reload a module in Python?

- importlib.reload function.
- import reload
- import logging
- none of the above

**Answer:** importlib.reload function.

**Explanation:** The `importlib.reload` function reloads a previously imported module in Python.
