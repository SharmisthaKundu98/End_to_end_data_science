```python
from sklearn import datasets

iris = datasets.load_iris()

import matplotlib.pyplot as plt
```

```python
iris
```

```python
import pandas as pd
import numpy as np

data1 = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
data1
```

```python
print("\nSummary Statistics by Category:")
print(data1['petal length (cm)'].describe())
```

```python
description_by_species = data1.groupby('target').describe()
```

```python
description_by_species
```

```python
petal_length = data1['petal length (cm)']
petal_length
```

```python
import seaborn as sns


plt.figure(figsize=(8, 6))
sns.histplot(petal_length, bins=20, color='skyblue', edgecolor='black',kde=True)
plt.title('Histogram of Petal Length for Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
skewness = petal_length.skew()
```
